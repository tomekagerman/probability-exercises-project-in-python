## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
#TODO

import random

# Set the number of iterations
iterations = 1000
# Counter for successful outcomes
success_count = 0

for _ in range(iterations):
    # Roll two 6-sided dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total_score = die1 + die2
    
    # Check if the sum is even OR greater than 7
    if total_score % 2 == 0 or total_score > 7:
        success_count = success_count + 1

# Calculate the estimated probability
probability = success_count / iterations

print(f"Total successes: {success_count}")
print(f"Estimated probability: {probability}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO

import random

iterations = 1000
success_scenario_1 = 0  # 3 White and 2 Red
success_scenario_2 = 0  # All same color

for _ in range(iterations):
    # Take 5 balls with replacement
    draw = []
    for _ in range(5):
        # Pick a random index (0-59) and get the color from the dictionary
        random_index = random.randint(0, 59)
        draw.append(ball_box[random_index])
    
    # Count colors in the current draw
    whites = draw.count('White')
    reds = draw.count('Red')
    greens = draw.count('Green')
    
    # 1. Check for exactly 3 White and 2 Red
    if whites == 3 and reds == 2:
        success_scenario_1 = success_scenario_1 + 1
        
    # 2. Check if all 5 balls are the same color
    if whites == 5 or reds == 5 or greens == 5:
        success_scenario_2 = success_scenario_2 + 1

# Calculate probabilities
prob_1 = success_scenario_1 / iterations
prob_2 = success_scenario_2 / iterations

print(f"Prob. 3 White and 2 Red: {prob_1}")
print(f"Prob. All same color: {prob_2}")
```
